### Problem 8:

<b> Find the top 5 Community Areas by average College Enrollment [number of students]. </b>

```python
%sql SELECT AVG(SAFETY_SCORE) FROM REPORT_CARDS
```

```python
%%sql
SELECT COMMUNITY_AREA_NAME, AVG(COLLEGE_ENROLLMENT) as "AVG_ENROLLMENT"
FROM REPORT_CARDS
GROUP BY COMMUNITY_AREA_NAME
ORDER BY AVG_ENROLLMENT DESC
LIMIT 5
```

### Problem 9:
<b> Use a sub-query to determine which Community Area has the least value for school Safety Score? </b>

```python
%%sql 
SELECT COMMUNITY_AREA_NAME FROM REPORT_CARDS
WHERE SAFETY_SCORE IN
(SELECT MIN(SAFETY_SCORE) FROM REPORT_CARDS)
```

### Problem 10:
<b> [Without using an explicit JOIN operator] Find the Per Capita Income of the Community Area which has a school Safety Score of 1. </b>

```python
%%sql
SELECT CD.PER_CAPITA_INCOME, RC.COMMUNITY_AREA_NAME, RC.SAFETY_SCORE
FROM CENSUS_DATA AS CD, REPORT_CARDS AS RC
WHERE RC.COMMUNITY_AREA_NUMBER=CD.COMMUNITY_AREA_NUMBER
ORDER BY RC.SAFETY_SCORE 
LIMIT 1
```
